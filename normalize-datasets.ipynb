```python
import pandas as pd
import numpy as np
```

```python
X = pd.read_csv('data/brain_counts.csv')
y = pd.read_csv('data/brain_metadata.csv')
```

```python
X
```

```python
classes = {}
ontology = y['cell_ontology_class']
labels = []

i = 0
for ont in ontology:
    
    if not ont in classes:
        classes[ont] = i
        i += 1
    labels.append(classes[ont])
```

```python
data = X.drop(columns=["Unnamed: 0", "zsGreen_transgene"])
```

```python
data['label'] = labels
```

```python
data
```

```python
data.to_csv('gene_expression.csv', index=False)
```
